```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read the data set to the python environment

```python
cars = pd.read_csv('cars_data.csv')
cars
```

# checking null values

```python
cars.isnull()
```

```python
cars.isnull().sum()  #checking total number  of null values
```

# Bar graph for males Vs Female buyers

```python
plt.figure(figsize=(7,7))
plt.bar(cars['Buyer Gender'],height=1, width= 0.4)
plt.title('MALE BUYERS Vs FEMALE BUYERS',fontsize= 20)
plt.yticks()
```

# Top 5 cars based on sales price

```python
top5car= cars.nlargest(5,['Sale Price'])
top5car = top5car['Make']+top5car['Model']
```

```python
print('Top 5 Cars based on sales price:\n \n',top5car)
```

# Least 5 cars based on their resale price

```python
least5cars= cars.nsmallest(5,['Resell Price'])
least5cars= least5cars['Make']+least5cars['Model']
print('Least 5  cars Based on Resell price:\n \n',least5cars)
```
